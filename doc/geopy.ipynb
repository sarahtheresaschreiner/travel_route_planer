# Trying out the options in geocoding

```python
from geopy.geocoders import Nominatim
from routingpy import Valhalla

geolocator = Nominatim(user_agent="my_geocoder")
```

```python
from pprint import pprint
```

```python
client = Valhalla(base_url="https://valhalla1.openstreetmap.de")
```

```python
def geocode_address(address):
    location = geolocator.geocode(address)
    if location:
        return [location.address, location.latitude, location.longitude]
    else:
        return None
```

```python
berlin = geocode_address("kurfürstendamm")
print(berlin)
""" hamburg = geocode_address("dietzenbach")
aachen = geocode_address("offenbach")
spots = [berlin, hamburg, aachen] """
```

```python
def getTeiler(n):
    ts=[1]
    for i in range(1, n):
        if n%i == 0:
            ts.append(i)
    return ts

getTeiler(10)
```

```python
matrix = client.matrix(locations=spots, profile='pedestrian')
pprint(matrix.durations)
```
