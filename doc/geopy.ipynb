# Trying out the options in geocoding

I use `geopy` to get the coordinats and `routingpy` with `Valhalla` to get a Matrix of all the distances (what will be the weight of the edges of my graph)

```python
from geopy.geocoders import Nominatim
from routingpy import Valhalla
from time import sleep

geolocator = Nominatim(user_agent="my_geocoder")
```

```python
from pprint import pprint
```

```python
client = Valhalla(base_url="https://valhalla1.openstreetmap.de")
```

```python
def geocode_address(address):
    location = geolocator.geocode(address)
    if location:
        return [location.latitude, location.longitude]
    else:
        return None
```

```python
langen = geocode_address("langen")
sleep(1)
dietzenbach = geocode_address("dietzenbach")
sleep(1)
offenbach = geocode_address("offenbach")
sleep(1)
spots = [langen, dietzenbach, offenbach]
```

```python
matrix = client.matrix(locations=spots, profile='auto')
pprint(matrix.durations)
```
